```python
from bs4 import BeautifulSoup


with open('../data/ck12.org/concepts/concepts-html/index.html') as html:
    soup = BeautifulSoup(html.read(), 'lxml')
```

```python
def remove_tag(soup, tag_properties):
    rem_divs = set()
    for tag in soup.find_all(*tag_properties):
        rem_divs.add(tag)
        tag.decompose()

    print('Removed {} elements'.format(len(rem_divs)))
    return soup
```

```python
#
# removing images
#
soup = remove_tag(['div', {'class':'x-ck12-img-postcard'}])
    
```

```python
def remove_tags(soup, tags):
    for tag in tags:
        print('Removing tag {}'.format(tag))
        soup = remove_tag(soup, tag)
    return soup
```

```python
TAGS = [['img'], ['div', {'class':'x-ck12-img-postcard'}]]#, ['table']]
soup = remove_tags(soup, TAGS)
```

```python
html_no_imgs = soup.prettify("utf-8")
with open("../data/ck12.org/concepts/concepts-html/index-noimgs.html", "wb") as file:
    file.write(html_no_imgs)
```

```python
from bs4 import NavigableString

SECTIONS = {'h1', 'h2', 'h3', 'h4', 'h5'}



def remove_sibling_tags(soup, content_filters=['Review', 'Explore More', 'References'], sections=SECTIONS):
    
    REM_TAGS = []
    for sec in sections:
        print('Processing for section', sec)
        for tag in soup.find_all(sec):
            remove_it = False
            for f in content_filters:
                # print('Processing filter', f)
                if tag.contents and f in tag.contents[0]:
                    remove_it = True
                    # print(tag.contents, len(list(tag.next_siblings)))
                    tags_to_remove = []
                    for sib_tag in tag.next_siblings:
                        # print('tag name', sib_tag.name, end= ' ')
                        if sib_tag.name in SECTIONS:
                            break
                        elif not isinstance(sib_tag, NavigableString):
                            # print('remove')
                            # sib_tag.decompose()
                            tags_to_remove.append(sib_tag)
                            REM_TAGS.append(sib_tag)
                        # else:
                            # print()
                    for rt in tags_to_remove:
                        rt.decompose()
            if remove_it:
                tag.decompose()
                REM_TAGS.append(tag)
    
    print(len(REM_TAGS))
    return soup
```

```python
def unwrap_links(soup):
    for tag in soup.find_all('a'):
        tag.unwrap()
    return soup
soup = unwrap_links(soup)
```

```python
from dataset import collect_document_paths
from preprocess import clean_text

import os

BOOK_PATHS = ['../data/ck12.org/books/CK-12-Biology-Advanced-Concepts/',
             '../data/ck12.org/books/CK-12-Biology-Concepts/',
             '../data/ck12.org/books/CK-12-Biology/',
             '../data/ck12.org/books/CK-12-Chemistry-Concepts-Intermediate/',
             '../data/ck12.org/books/CK-12-Earth-Science-Concepts-For-High-School/',
             '../data/ck12.org/books/CK-12-Earth-Science-Concepts-For-Middle-School/',
             '../data/ck12.org/books/CK-12-Earth-Science-For-Middle-School/',
             '../data/ck12.org/books/CK-12-Life-Science-Concepts-For-Middle-School/',
             '../data/ck12.org/books/CK-12-Life-Science-For-Middle-School/',
             '../data/ck12.org/books/CK-12-Physical-Science-Concepts-For-Middle-School/',
             '../data/ck12.org/books/CK-12-Physics---Intermediate/',
             '../data/ck12.org/books/CK-12-Physics-Concepts-Intermediate/',
             '../data/ck12.org/books/Human-Biology-Ecology/',
             '../data/ck12.org/books/Peoples-Physics-Concepts/']

CONTENTS_EXT = '.contents'

all_contents = []

for book in BOOK_PATHS:
    print('Processing book', book)
    book_contents = []
    for doc in collect_document_paths(book, '.txt'):
        print('Processing', doc)
        with open(doc, 'r') as chapter:
            all_lines =  chapter.readlines()
            # print('there are ', len(all_lines))
            chapter_lines = [clean_text(l) for l in all_lines]
            # print('there are ', len(all_lines))
            book_contents.extend(chapter_lines)
    print(len(book_contents))
            
    all_contents.extend(book_contents)
    
    #
    # writing into a single file
    book_name = os.path.basename(book[:-1]) 
    book_contents_path = os.path.join(book, book_name + CONTENTS_EXT)
    print('Saving to ', book_name, book_contents_path)
    with open(book_contents_path, 'w') as bc:
        for line in book_contents:
            bc.write(line)
          

with open('../data/science-books.contents.txt', 'w') as bc:
    for line in all_contents:
        bc.write(line)
```

```python
def concatenate_files(filename_list, output_path):
    with open(output_path, 'w') as outfile:
        for fname in filename_list:
            with open(fname, 'r') as infile:
                for line in infile:
                    outfile.write(line)
                
```
